```sql
-- NULL AUDIT — check for missing values across key tables
-- Focus on primary keys and critical columns

SELECT
    'drivers' AS tbl,
    SUM(CASE WHEN DriverID IS NULL THEN 1 ELSE 0 END)        AS null_pk,
    SUM(CASE WHEN Rating IS NULL THEN 1 ELSE 0 END)          AS null_rating,
    SUM(CASE WHEN Age IS NULL THEN 1 ELSE 0 END)             AS null_age
FROM drivers
UNION ALL
SELECT
    'passengers',
    SUM(CASE WHEN PassengerID IS NULL THEN 1 ELSE 0 END),
    SUM(CASE WHEN MemberCategory IS NULL THEN 1 ELSE 0 END),
    SUM(CASE WHEN Age IS NULL THEN 1 ELSE 0 END)
FROM passengers
UNION ALL
SELECT
    'booking',
    SUM(CASE WHEN BookingID IS NULL THEN 1 ELSE 0 END),
    SUM(CASE WHEN Fare IS NULL THEN 1 ELSE 0 END),
    SUM(CASE WHEN Distance IS NULL THEN 1 ELSE 0 END)
FROM booking;
```
