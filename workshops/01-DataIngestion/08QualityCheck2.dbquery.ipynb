```sql
-- DUPLICATE CHECK — verify primary key uniqueness

SELECT 'taxi_type' AS tbl, COUNT(*) AS total, COUNT(DISTINCT TaxiTypeID) AS distinct_pk, COUNT(*) - COUNT(DISTINCT TaxiTypeID) AS duplicates FROM taxi_type
UNION ALL
SELECT 'payment_type',        COUNT(*),          COUNT(DISTINCT PaymentTypeID),              COUNT(*) - COUNT(DISTINCT PaymentTypeID) FROM payment_type
UNION ALL
SELECT 'place',               COUNT(*),          COUNT(DISTINCT PlaceID),                    COUNT(*) - COUNT(DISTINCT PlaceID) FROM place
UNION ALL
SELECT 'drivers',             COUNT(*),          COUNT(DISTINCT DriverID),                   COUNT(*) - COUNT(DISTINCT DriverID) FROM drivers
UNION ALL
SELECT 'taxi',                COUNT(*),          COUNT(DISTINCT TaxiID),                     COUNT(*) - COUNT(DISTINCT TaxiID) FROM taxi
UNION ALL
SELECT 'passengers',          COUNT(*),          COUNT(DISTINCT PassengerID),                COUNT(*) - COUNT(DISTINCT PassengerID) FROM passengers
UNION ALL
SELECT 'booking',       COUNT(*),          COUNT(DISTINCT BookingID),                  COUNT(*) - COUNT(DISTINCT BookingID) FROM booking
ORDER BY tbl;
```
