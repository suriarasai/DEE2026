```sql
-- MASTER ROW COUNT — all 10 tables in one query
-- Confirms every CSV was loaded with the expected row count

SELECT 'taxi_type'     AS table_name, COUNT(*) AS rows FROM taxi_type
UNION ALL
SELECT 'payment_type',                COUNT(*)         FROM payment_type
UNION ALL
SELECT 'place',                       COUNT(*)         FROM place
UNION ALL
SELECT 'drivers',                     COUNT(*)         FROM drivers
UNION ALL
SELECT 'taxi',                        COUNT(*)         FROM taxi
UNION ALL
SELECT 'passengers',                  COUNT(*)         FROM passengers
UNION ALL
SELECT 'rewards',                     COUNT(*)         FROM rewards
UNION ALL
SELECT 'payment',                     COUNT(*)         FROM payment
UNION ALL
SELECT 'booking',                     COUNT(*)         FROM booking
ORDER BY rows;
```
