```sql
-- REFERENTIAL INTEGRITY — check foreign keys resolve correctly
-- An orphan is a FK value in the child table with no match in
-- the parent table. Orphans indicate data quality problems.


-- Check 1: Do all Taxi.TaxiTypeID values exist in taxi_type?
SELECT 'taxi → taxi_type' AS fk_check,
       COUNT(*) AS orphan_count
FROM taxi t
LEFT JOIN taxi_type tt ON t.TaxiTypeID = tt.TaxiTypeID
WHERE tt.TaxiTypeID IS NULL

UNION ALL

-- Check 2: Do all booking.DriverID values exist in drivers?
SELECT 'booking → drivers',
       COUNT(*)
FROM booking b
LEFT JOIN drivers d ON b.DriverID = d.DriverID
WHERE d.DriverID IS NULL

UNION ALL

-- Check 3: Do all booking.TaxiID values exist in taxi?
SELECT 'booking → taxi',
       COUNT(*)
FROM booking b
LEFT JOIN taxi t ON b.TaxiID = t.TaxiID
WHERE t.TaxiID IS NULL

UNION ALL

-- Check 4: Do all booking.PickupPlaceID values exist in place?
SELECT 'booking → place (pickup)',
       COUNT(*)
FROM booking b
LEFT JOIN place p ON b.PickupPlaceID = p.PlaceID
WHERE p.PlaceID IS NULL;
```
